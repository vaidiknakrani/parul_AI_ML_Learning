```python
import tensorflow as tf

# Define sentences to tokenize
sentences = [
    "I love travelling",
    "Benagluru is a nice city",
    "I Love Bengaluru",
    "Lal Bagh is in benagluru"
]

# Initialize the Tokenizer from TensorFlow Keras
tokenizer = tf.keras.preprocessing.text.Tokenizer()

# Fit the tokenizer on the provided sentences
tokenizer.fit_on_texts(sentences)

# Convert sentences to sequences of tokens
sequences = tokenizer.texts_to_sequences(sentences)

# Print the word index mapping
print("Word Index:\n", tokenizer.word_index)

# Print token sequences
for i, sequence in enumerate(sequences):
    print(f"Sentence {i+1} tokens:", sequence)
```
